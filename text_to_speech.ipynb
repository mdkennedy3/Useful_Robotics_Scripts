```python
#@title üìù  Text-to-Speech ‚Üí WAV
from gtts import gTTS
from pydub import AudioSegment
from IPython.display import Audio, display
from google.colab import files

# 1) Get input text
text = input("Enter the text you want to speak: ").strip()
if not text:
    raise ValueError("No text provided!")

# 2) Generate an MP3 with gTTS
tts = gTTS(text, lang='en')
tts.save('output.mp3')

# 3) Convert MP3 ‚Üí WAV
audio = AudioSegment.from_mp3('output.mp3')
audio.export('output.wav', format='wav')

# 4) Play inline
print("‚ñ∂Ô∏è Playing back your TTS result:")
display(Audio('output.wav', autoplay=True))

# 5) Download link
files.download('output.wav')
```

```python
#@title üìù Text-to-Speech ‚Üí WAV with a male voice (fixed)
import nest_asyncio
nest_asyncio.apply()

import asyncio
import edge_tts
from pydub import AudioSegment
from IPython.display import Audio, display
from google.colab import files

async def synthesize(text, voice, mp3_path):
    communicate = edge_tts.Communicate(text, voice)
    await communicate.save(mp3_path)

# 1) Get input text
text = input("Enter the text you want to speak: ").strip()
if not text:
    raise ValueError("No text provided!")

# 2) Generate MP3 using edge-tts with a male voice
mp3_file = "output_male.mp3"
voice_name = "en-US-GuyNeural"   # male voice

loop = asyncio.get_event_loop()
loop.run_until_complete(synthesize(text, voice_name, mp3_file))

# 3) Convert MP3 ‚Üí WAV
audio = AudioSegment.from_mp3(mp3_file)
wav_file = "output_male.wav"
audio.export(wav_file, format="wav")

# 4) Play inline
print("‚ñ∂Ô∏è Playing back with ‚ÄúGuyNeural‚Äù voice:")
display(Audio(wav_file, autoplay=True))

# 5) Download link
files.download(wav_file)
```
